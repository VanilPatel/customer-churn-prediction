Exploratory Data Analysis on Telecom Customer Churn Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
print('✅ Libraries loaded!')
```

```python
# Generate synthetic telecom dataset
np.random.seed(42)
n = 7043
df = pd.DataFrame({
    'customerID': [f'CUST-{i:05d}' for i in range(n)],
    'gender': np.random.choice(['Male', 'Female'], n),
    'SeniorCitizen': np.random.choice([0, 1], n, p=[0.84, 0.16]),
    'tenure': np.random.randint(0, 72, n),
    'Contract': np.random.choice(['Month-to-month', 'One year', 'Two year'], n, p=[0.55, 0.24, 0.21]),
    'InternetService': np.random.choice(['DSL', 'Fiber optic', 'No'], n, p=[0.34, 0.44, 0.22]),
    'MonthlyCharges': np.random.uniform(18, 119, n).round(2),
    'TotalCharges': np.random.uniform(18, 8685, n).round(2),
    'PaymentMethod': np.random.choice(['Electronic check', 'Mailed check', 'Bank transfer', 'Credit card'], n),
    'Churn': np.random.choice(['Yes', 'No'], n, p=[0.265, 0.735])
})

print(f'Dataset shape: {df.shape}')
print(f'Churn rate: {(df.Churn=="Yes").mean():.1%}')
df.head()
```

```python
# ── Churn Distribution ──
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

churn_counts = df['Churn'].value_counts()
axes[0].pie(churn_counts, labels=['No Churn', 'Churn'], autopct='%1.1f%%',
            colors=['#2196F3', '#F44336'], startangle=90)
axes[0].set_title('Churn Distribution', fontsize=14, fontweight='bold')

sns.countplot(x='Contract', hue='Churn', data=df, palette=['#2196F3', '#F44336'], ax=axes[1])
axes[1].set_title('Churn by Contract Type', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Contract Type')

plt.tight_layout()
plt.savefig('churn_distribution.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# ── Tenure vs Monthly Charges by Churn ──
plt.figure(figsize=(10, 5))
colors = {'Yes': '#F44336', 'No': '#2196F3'}
for churn_val, group in df.groupby('Churn'):
    plt.scatter(group['tenure'], group['MonthlyCharges'],
                alpha=0.3, label=f'Churn={churn_val}', color=colors[churn_val], s=15)
plt.xlabel('Tenure (months)', fontsize=12)
plt.ylabel('Monthly Charges ($)', fontsize=12)
plt.title('Tenure vs Monthly Charges by Churn Status', fontsize=14, fontweight='bold')
plt.legend()
plt.savefig('tenure_charges_scatter.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# ── Correlation Heatmap ──
numeric_cols = ['SeniorCitizen', 'tenure', 'MonthlyCharges', 'TotalCharges']
df_num = df[numeric_cols].copy()
df_num['Churn_binary'] = (df['Churn'] == 'Yes').astype(int)

plt.figure(figsize=(7, 5))
sns.heatmap(df_num.corr(), annot=True, fmt='.2f', cmap='coolwarm',
            linewidths=0.5, square=True)
plt.title('Feature Correlation Heatmap', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('correlation_heatmap.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# ── Summary Stats ──
print('=== Dataset Summary ===')
print(df.describe().round(2))
print('\n=== Missing Values ===')
print(df.isnull().sum())
print('\n=== Churn Rate by Contract ===')
print(df.groupby('Contract')['Churn'].apply(lambda x: (x=='Yes').mean()).round(3))
```
